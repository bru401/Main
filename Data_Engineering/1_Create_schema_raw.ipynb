# Creating Schema
We'll start by creating an empty database.

```python
%sql
CREATE DATABASE IF NOT EXISTS bronze;
```

Then, we create the empty schema that will receive the raw data, currently stored in this very same GitHub folder. This will make for the bronze layer of this database. Through this schema, we define the first constraints for the dataset, namely the type constraints for each column.

```python
%sql
CREATE TABLE IF NOT EXISTS bronze.raw_data (
  ID BIGINT,
  Year_Birth BIGINT,
  Education STRING,
  Marital_Status STRING,
  Income BIGINT,
  Kidhome BIGINT,
  Teenhome BIGINT,
  Dt_Customer DATE,
  Recency BIGINT,
  MntWines BIGINT,
  MntFruits BIGINT,
  MntMeatProducts BIGINT,
  MntFishProducts BIGINT,
  MntSweetProducts BIGINT,
  MntGoldProds BIGINT,
  NumDealsPurchases BIGINT,
  NumWebPurchases BIGINT,
  NumCatalogPurchases BIGINT,
  NumStorePurchases BIGINT,
  NumWebVisitsMonth BIGINT,
  AcceptedCmp3 BIGINT,
  AcceptedCmp4 BIGINT,
  AcceptedCmp5 BIGINT,
  AcceptedCmp1 BIGINT,
  AcceptedCmp2 BIGINT,
  Complain BIGINT,
  Z_CostContact BIGINT,
  Z_Revenue BIGINT,
  Response BIGINT
)
```
